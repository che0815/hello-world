```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
NN001 = np.genfromtxt('NN.csv', delimiter=',')
NN002 = np.genfromtxt('NN002.csv', delimiter=',')
NN0015= np.genfromtxt('NN0015.csv', delimiter=',')
NN0005= np.genfromtxt('NN0005.csv', delimiter=',')
HL150 = np.genfromtxt('HL150.csv', delimiter=',')
HL250 = np.genfromtxt('HL250.csv', delimiter=',')
```

```python
def getepochtest(a):
    b = np.delete(a, np.where((a-np.around(a, decimals=2)) == 0)[0])
    return b
```

```python
NN001_  = getepochtest(NN001)
NN002_  = getepochtest(NN002)
NN0015_ = getepochtest(NN0015)
NN0005_ = getepochtest(NN0005)
HL150_  = getepochtest(HL150)
HL250_  = getepochtest(HL250)
```

```python
plt.figure(num=None, figsize=(9, 5), facecolor='w', edgecolor='k')
plt.subplot(1,2,1)
plt.plot(np.arange(0,len(NN0005_)),NN0005_,'.-',label='Learning Rate: 0.005', alpha=0.8)
plt.plot(np.arange(0,len(NN001_)) ,NN001_ ,'.-',label='Learning Rate: 0.01' , alpha=0.3)
plt.plot(np.arange(0,len(NN0015_)),NN0015_,'.-',label='Learning Rate: 0.015', alpha=0.3)
plt.plot(np.arange(0,len(NN002_)) ,NN002_ ,'.-',label='Learning Rate: 0.02' , alpha=0.3)
plt.plot(np.arange(0,len(HL150_)) ,HL150_ ,'.-',label='Hidden N 150'        , alpha=0.3)
plt.plot(np.arange(0,len(HL250_)) ,HL250_ ,'.-',label='Hidden N 250'        , alpha=0.8)

plt.legend(loc='lower right')
plt.xlabel('Epochs')
plt.ylabel('Performance')

plt.subplot(1,2,2)
plt.plot(np.arange(0,len(np.ediff1d(NN0005_))),np.ediff1d(NN0005_),'.-',label='Learning Rate: 0.005', alpha=0.8)
plt.plot(np.arange(0,len(np.ediff1d(NN001_))) ,np.ediff1d(NN001_) ,'.-',label='Learning Rate: 0.01' , alpha=0.3)
plt.plot(np.arange(0,len(np.ediff1d(NN0015_))),np.ediff1d(NN0015_),'.-',label='Learning Rate: 0.015', alpha=0.3)
plt.plot(np.arange(0,len(np.ediff1d(NN002_))) ,np.ediff1d(NN002_) ,'.-',label='Learning Rate: 0.02' , alpha=0.3)
plt.plot(np.arange(0,len(np.ediff1d(HL150_))) ,np.ediff1d(HL150_) ,'.-',label='Hidden N 150'        , alpha=0.3)
plt.plot(np.arange(0,len(np.ediff1d(HL250_))) ,np.ediff1d(HL250_) ,'.-',label='Hidden N 250'        , alpha=0.8)
```
